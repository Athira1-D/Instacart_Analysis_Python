```python
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
import seaborn as sns
import scipy
```

```python
high_activity_customers = pd.read_pickle(os.path.join(path, '02 Data', 'Prepared Data', 'high_acti_customer_ords_prods_merged.pkl'))
```

```python
#Checking dependent profile with spending_flag
spending_depedent = pd.crosstab(high_activity_customers['spending_flag'],high_activity_customers['number_of_dependents'],dropna=False)
```

```python
spending_depedent_bar = spending_depedent.plot.bar()
```

```python
#Checking dependent profile with income flag ang age flag
income_age = pd.crosstab(high_activity_customers['Income'],high_activity_customers['age_flag'],dropna=False)
```

```python
income_age_bar = income_age.plot.bar()
```

```python
spending_age = pd.crosstab(high_activity_customers['spending_flag'],high_activity_customers['age_flag'],dropna=False)
```

```python
spending_age_bar = spending_age.plot.bar()
```

```python
profile_spending = high_activity_customers .groupby(['spending_flag', 'customer_profile']).size().unstack(fill_value=0)
```

```python
profile_spending_bar = profile_spending.plot.bar()
```
