```python
#Define a function
def price_label(row):

  if row['prices'] <= 5:
    return 'Low-range product'
  elif (row['prices'] > 5) and (row['prices'] <= 15):
    return 'Mid-range product'
  elif row['prices'] > 15:
    return 'High range'
  else: return 'Not enough data'
```

```python
df['price_range'] = df.apply(price_label, axis=1)
```

```python
df['price_range'].value_counts(dropna=False)
```

```python
df['prices'].max()
```

```python
df.loc[df['prices'] > 15, 'price_range_loc'] = 'High-range product'
```

```python
df.loc[(df['prices'] <= 15) & (df['prices'] > 5), 'price_range_loc'] = 'Mid-range product' 
```

```python
df.loc[df['prices'] <= 5, 'price_range_loc'] = 'Low-range product'
```

```python
#Finding which day most orders take place
df_ords_prods_merge['orders_day_of_week'].value_counts(dropna=False)
```

```python
#Grouping the day most orders take place
result = []
for value in df_ords_prods_merge["orders_day_of_week"]:
  if value == 0:
    result.append("Busiest day")
  elif value == 4:
    result.append("Least busy")
  else:
    result.append("Regularly busy")
```

```python
result
```

```python
#Creating new column
df_ords_prods_merge['Busiest day']=result
```

```python
#Frequency of column
df_ords_prods_merge['Busiest day'].value_counts(dropna=False)
```

```python
#“Busiest day” to become “Busiest days” (plural)
result = []
for value in df_ords_prods_merge["orders_day_of_week"]:
  if value == 0 or value==1:
    result.append("Busiest days")
  elif value == 4 or value==3:
    result.append("Slowest days")
  else:
    result.append("Regularly busy")
```

```python
#Creating new column
df_ords_prods_merge['Busiest days']=result
```

```python
#Frequency of column 
df_ords_prods_merge['Busiest days'].value_counts(dropna=False)
```

```python
#Finding Busiest hours of the day
df_ords_prods_merge['order_hour_of_day'].value_counts(dropna=False)
```

```python
#Grouping by Busiest hours of the day
result1 = []
for value1 in df_ords_prods_merge["order_hour_of_day"]:
  if value1 == 10 or value1==11 or value1 == 14 or value1==15 or value1 == 13 or value1==12 or value1 == 16 or value1==9:
    result1.append("Most orders")
  elif value1 == 17 or value1==8 or value1 == 18 or value1==19 or value1 == 20 or value1==7 or value1 == 21 or value1==22:
    result1.append("Average orders")
  else:
    result1.append("Fewest orders")
```

```python
#Creating new column named 'busiest_period_of_day'
df_ords_prods_merge['busiest_period_of_day']=result1
```

```python
#Frequency of column 'busiest_period_of_day'
df_ords_prods_merge['busiest_period_of_day'].value_counts(dropna=False)
```
